```python
from scapy.all import rdpcap, Ether, IP, TCP

# Load the packet capture file
pkt = rdpcap("dump.pcap")
pkt.show()
print("")

# Iterate through each packet in the capture
for index, packet in enumerate(pkt):
    # Extract Ethernet, IP, and TCP layers
    if Ether in packet and IP in packet and TCP in packet:
        ether_frame = packet[Ether]
        ip_packet = packet[IP]
        tcp_segment = packet[TCP]

        # Extract TCP flags
        tcp_flags = tcp_segment.flags

        # Display relevant information
        print(f"Packet {index + 1}:")
        print(f"Source IP: {ip_packet.src}, Destination IP: {ip_packet.dst}")
        print(f"Source Port: {tcp_segment.sport}, Destination Port: {tcp_segment.dport}")
        print(f"TCP Flags: {tcp_flags}")
        print("")

# Add more analysis or processing as needed
```
